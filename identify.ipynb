```python
VAR_ID = "81"
START_TH_ID = 100
END_TH_ID = 125

def explore_var(VAR_ID, START_TH_ID = 100, END_TH_ID = 125):
    URL = "https://webapi.bps.go.id/v1/api/list/model/data/domain/{domain}/var/{var}/th/{th}/key/{key}/"
    result_explore = []
    
    print(f"Memulai penjelajahan data untuk VAR ID: {VAR_ID} dari tahun ID: {START_TH_ID} sampai {END_TH_ID}")
    
    # Looping dengan loncatan 2 untuk setiap panggilan API 
    for th_start in range(START_TH_ID, END_TH_ID + 1, 2):
        th_end = th_start + 1
        if th_end > END_TH_ID:
            th_end = END_TH_ID
        
        rentang_th_str = f"{th_start}:{th_end}"
        if th_start == th_end:
            rentang_th_str = str(th_start)
            
        print(f" > Memanggil API untuk rentang ID tahun: {rentang_th_str}...")
        url = URL.format(domain=DOMAIN, var = VAR_ID, th=rentang_th_str, key = API_KEY)
        
        try:
            response = requests.get(url, timeout=20)
            
            if response.status_code != 200:
                print(f"   ! Panggian API gagal, Status: {response.status_code}.")
                for th_id in range(th_start, th_end+1):
                    result_explore.append({'id_tahun':th_id, 'status':'Error API'})
                continue
            
            data_json = response.json()
            
            if data_json.get("data-availability") != "available":
                print("   - Data tidak tersedia untuk rentang ini.")
                for th_id in range(th_start, th_end + 1):
                    result_explore.append({'id_tahun': th_id, 'status':'Data Tidak Ditemukan'})
                    time.sleep(1)
                    continue
            
            # --- 1A. Ekstrak Semua Metadata Umum -- 
            # metadata_umum = {
            #     'last_update': data_json.get('last_update'),
            #     'subject_id': data_json['subject'][0]['val'],
            #     'subject_label': data_json['subject'][0]['label'],
            #     'var_id' : data_json['var'][0]['var'],
            #     'var_label': data_json['var'][0]['label'],
            #     'var_unit': data_json['var'][0]['unit'],
            #     'var_decimal': data_json['var'][0]['decimal'],
            #     'turvar_id': data_json['turvar'][0]['val'],
            #     'turvar_label': data_json['turvar'][0]['label'],
            #     'vervar_id': data_json['vervar'][0]['val'],
            #     'vervar_label': data_json['vervar'][0]['label'],
            #     'turtahun_id': data_json['turtahun'][0]['val'],
            #     'turtahun_label': data_json['turtahun'][0]['label']
            # }
            
            metadata_umum = {
                'last_update': data_json.get('last_update'),
                # Cek dulu apakah kunci 'subject' ada, baru ambil nilainya. Jika tidak, beri nilai None.
                'subject_id': data_json.get('subject')[0]['val'] if data_json.get('subject') else None,
                'subject_label': data_json.get('subject')[0]['label'] if data_json.get('subject') else None,
                
                # Terapkan pola yang sama untuk kunci lainnya
                'var_id': data_json.get('var')[0]['val'] if data_json.get('var') else None,
                'var_label': data_json.get('var')[0]['label'] if data_json.get('var') else None,
                'var_unit': data_json.get('var')[0]['unit'] if data_json.get('var') else None,
                'var_decimal': data_json.get('var')[0]['decimal'] if data_json.get('var') else None,
                'turvar_id': data_json.get('turvar')[0]['val'] if data_json.get('turvar') else None,
                'turvar_label': data_json.get('turvar')[0]['label'] if data_json.get('turvar') else None,
                'vervar_id': data_json.get('vervar')[0]['val'] if data_json.get('vervar') else None,
                'vervar_label': data_json.get('vervar')[0]['label'] if data_json.get('vervar') else None,
                'turtahun_id': data_json.get('turtahun')[0]['val'] if data_json.get('turtahun') else None,
                'turtahun_label': data_json.get('turtahun')[0]['label'] if data_json.get('turtahun') else None
                }
            
            
            # --- 1B. Ekstrak dan gabungkan Metadata 'related' ---
            # Mengambil semua related, id, title, dan link
            related_items = data_json.get('related', [])
            list_related_id = [str(item.get('id', '')) for item in related_items]
            list_related_title = [str(item.get('title', '')) for item in related_items]
            list_related_link = [str(item.get('link', '')) for item in related_items]
            
            # Gabungkan list menjadi satu string dengan pemisah ;
            metadata_umum['related_ids']= "; ".join(list_related_id)
            metadata_umum['related_titles']= "; ".join(list_related_title)
            metadata_umum['related_links']= "; ".join(list_related_link)
            
            # tahun_lookup = {item['val']: item['label'] for item in data_json['tahun']}
            # Ambil list 'tahun' dengan aman. Jika tidak ada, gunakan list kosong [].
            tahun_list = data_json.get('tahun', [])
            # Buat dictionary lookup dari list tersebut (jika listnya kosong, hasilnya akan jadi dictionary kosong)
            tahun_lookup = {item['val']: item['label'] for item in tahun_list}
            
            #--- 2. Periksa Setiap Tahun dalam Rentang yang diminta ---
            for current_th_id in range(th_start, th_end + 1):
                kunci = f"{metadata_umum['vervar_id']}{metadata_umum['var_id']}{metadata_umum['turvar_id']}{current_th_id}{metadata_umum['turtahun_id']}"
                # nilai = data_json['datacontent'].get(kunci)
                nilai = data_json.get('datacontent', {}).get(kunci)
                
                baris_data = metadata_umum.copy()
                baris_data['id_tahun'] = current_th_id
                baris_data['tahun'] = tahun_lookup.get(current_th_id)
                baris_data['nilai'] = nilai
                
                if nilai is not None:
                    baris_data['status'] = 'Data Tersedia'
                    print(f"    âœ“ Data ditemukan untuk ID tahun {current_th_id} ({baris_data['tahun']})")
                else:
                    baris_data['status'] = 'Data Tidak Ditemukan'
                    print(f"    âœ— Data tidak ditemukan untuk ID tahun {current_th_id}")
                
                # Memasukkan baris yang sudah berisi semua data ke hasil akhir
                result_explore.append(baris_data)
                
        except requests.exceptions.RequestException as e:
            print(f"   !Terjadi error koneksi: {e}")
            for th_id in range(th_start, th_end + 1):
                result_explore.append({'id_tahun': th_id, 'status':'Error Koneksi'})
        
        time.sleep(1)
    print("\n---Proses Penjelajahan Selesai ---")
    
    # Mengubah hasil menjadi tabel pandas 
    df_hasil = pd.DataFrame(result_explore)
    
    # Mengatur urutan kolom agar lebih mudah dibaca dan memastikan semua kolom ada
    kolom_utama = ['id_tahun', 'tahun', 'nilai', 'status', 'last_update', 'var_label', 'related_ids', 'related_titles', 'related_links']
    semua_kolom = kolom_utama + [kolom for kolom in df_hasil.columns if kolom not in kolom_utama]
    df_hasil = df_hasil[semua_kolom]
    
    df_hasil = df_hasil.sort_values(by='id_tahun').reset_index(drop=True)
    
    nama_file = f"assets/master/explore_data_var_{VAR_ID}.csv"
    df_hasil.to_csv(nama_file)
    
    print(f"Data Variabel {VAR_ID} Berhasil di Simpan pada {nama_file}")
    
    return(df_hasil)
            
```

```python
# Ketika rownya di tabel lebih dari 1
def explore_var2(VAR_ID, START_TH_ID=100, END_TH_ID=125):
    URL = "https://webapi.bps.go.id/v1/api/list/model/data/domain/{domain}/var/{var}/th/{th}/key/{key}/"
    result_explore = []
    
    print(f"Memulai penjelajahan data untuk VAR ID: {VAR_ID} dari tahun ID: {START_TH_ID} sampai {END_TH_ID}")
    
    # Looping dengan loncatan 2 untuk setiap panggilan API 
    for th_start in range(START_TH_ID, END_TH_ID + 1, 2):
        th_end = th_start + 1
        if th_end > END_TH_ID:
            th_end = th_end
        
        rentang_th_str = f"{th_start}:{th_end}"
        if th_start == th_end:
            rentang_th_str = str(th_start)
            
        print(f" > Memanggil API var {VAR_ID} untuk rentang ID tahun: {rentang_th_str}...")
        url = URL.format(domain=DOMAIN, var=VAR_ID, th=rentang_th_str, key=API_KEY)
        
        try:
            response = requests.get(url, timeout=20)
            
            if response.status_code != 200:
                print(f"   ! Panggilan API var {VAR_ID} gagal, Status: {response.status_code}.")
                for th_id in range(th_start, th_end + 1):
                    result_explore.append({'id_tahun': th_id, 'status': 'Error API'})
                continue
            
            data_json = response.json()
            
            if data_json.get("data-availability") != "available":
                print(f"   - Data var {VAR_ID} tidak tersedia untuk rentang ini.")
                for th_id in range(th_start, th_end + 1):
                    result_explore.append({'id_tahun': th_id, 'status': 'Data Tidak Ditemukan'})
                time.sleep(1)
                continue
            
            # --- 1A. Ekstrak Metadata yang Umum untuk Semua Baris dalam Respons Ini ---
            metadata_umum = {
                'last_update': data_json.get('last_update'),
                'subject_id': data_json.get('subject')[0]['val'] if data_json.get('subject') else None,
                'subject_label': data_json.get('subject')[0]['label'] if data_json.get('subject') else None,
                'var_id': data_json.get('var')[0]['val'] if data_json.get('var') else None,
                'var_label': data_json.get('var')[0]['label'] if data_json.get('var') else None,
                'var_unit': data_json.get('var')[0]['unit'] if data_json.get('var') else None,
            }
            
            # --- 1B. Ekstrak 'related' ---
            related_items = data_json.get('related', [])
            metadata_umum['related_ids'] = "; ".join([str(item.get('id', '')) for item in related_items])
            metadata_umum['related_titles'] = "; ".join([str(item.get('title', '')) for item in related_items])
            metadata_umum['related_links'] = "; ".join([str(item.get('link', '')) for item in related_items])

            # --- 2. Lakukan Loop Bertingkat untuk Vervar, Turvar, dan Tahun ---
            vervar_list = data_json.get('vervar', [])
            turvar_list = data_json.get('turvar', [])
            turtahun_list = data_json.get('turtahun', [])
            tahun_list = data_json.get('tahun', [])
            tahun_lookup = {item['val']: item['label'] for item in tahun_list}
            
            # Loop untuk setiap item di VERVAR
            for vervar_item in vervar_list:
                # Loop untuk setiap item di TURVAR
                for turvar_item in turvar_list:
                    # Loop untuk setiap TAHUN yang diminta dalam rentang ini
                    for current_th_id in range(th_start, th_end + 1):
                        
                        id_turtahun = turtahun_list[0]['val'] # Asumsi turtahun selalu ada 1
                        
                        # --- 3. Bentuk Kunci dan Cari Nilai ---
                        kunci = f"{vervar_item['val']}{metadata_umum['var_id']}{turvar_item['val']}{current_th_id}{id_turtahun}"
                        nilai = data_json.get('datacontent', {}).get(kunci)
                        
                        # --- 4. Buat Baris Data Lengkap ---
                        baris_data = metadata_umum.copy()
                        baris_data.update({
                            'vervar_id': vervar_item['val'],
                            'vervar_label': vervar_item['label'],
                            'turvar_id': turvar_item['val'],
                            'turvar_label': turvar_item['label'],
                            'id_tahun': current_th_id,
                            'tahun': tahun_lookup.get(current_th_id),
                            'nilai': nilai
                        })
                        
                        if nilai is not None:
                            baris_data['status'] = 'Data Tersedia'
                            print(f"    âœ“ Data ditemukan untuk: '{vervar_item['label']}' tahun {tahun_lookup.get(current_th_id)}")
                        else:
                            baris_data['status'] = 'Data Tidak Ditemukan'
                        
                        result_explore.append(baris_data)
                        
        except requests.exceptions.RequestException as e:
            print(f"   ! Terjadi error koneksi: {e}")
            for th_id in range(th_start, th_end + 1):
                result_explore.append({'id_tahun': th_id, 'status': 'Error Koneksi'})
        
        time.sleep(1)
        
    print(f"\n--- Proses Penjelajahan var {VAR_ID} Selesai ---")
    
    df_hasil = pd.DataFrame(result_explore)
    
    # Atur urutan kolom
    kolom_utama = ['id_tahun', 'tahun', 'vervar_label', 'nilai', 'status', 'last_update', 'var_label']
    semua_kolom = kolom_utama + [kolom for kolom in df_hasil.columns if kolom not in kolom_utama]
    df_hasil = df_hasil[semua_kolom]
    
    df_hasil = df_hasil.sort_values(by=['id_tahun', 'vervar_id']).reset_index(drop=True)
    
    nama_file = f"assets/master/explore_data2_var_{VAR_ID}.csv"
    df_hasil.to_csv(nama_file, index=False)
    
    print(f"Data Variabel {VAR_ID} Berhasil di Simpan pada {nama_file}")
    
    return df_hasil
```
